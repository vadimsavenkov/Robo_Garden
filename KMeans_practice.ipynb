```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
from sklearn.datasets import make_blobs
```

```python
#create data with 4 clusters and labels
data = make_blobs(n_samples=1000,
    n_features=2, centers=4,
    cluster_std=3.0, random_state=42)
```

```python
#check how does the data look like
data
```

```python
data[0].shape
```

```python
#create scatter plot to visualize clusters
plt.scatter(x = data[0][:, 0], y = data[0][:,1] )
```

```python
plt.scatter(x = data[0][:, 0], y = data[0][:,1], c = data[1], cmap = 'rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
#pass 4 clusters to imported kmeans model and create the model
kmeans = KMeans(n_clusters = 4)
```

```python
#train our model
kmeans.fit(data[0])
```

```python
#create cluster centroid
cent = kmeans.cluster_centers_
cent
```

```python
#check labels on our trained model
kmeans.labels_
```

```python
#create 2 scatter plots with predicted labels on the left 
#and original labels on the right. Put centroids on the plot 
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,sharey = True, figsize = (10,6))

ax1.scatter(x = data[0][:, 0], y = data[0][:,1], c = kmeans.labels_, cmap = 'rainbow')
ax1.scatter(x=cent[:, 0], y=cent[:,1], s = 200, c = 'black')

ax2.scatter(x = data[0][:, 0], y = data[0][:,1], c = data[1], cmap = 'rainbow')
```

```python
#use the range function and pass 9 k_values to create elbow method
sum_sq = {}
for k in range(1,10):
    kmeans = KMeans(n_clusters = k).fit(data[0])
    sum_sq[k] = kmeans.inertia_
```

```python
sum_sq
```

```python

plt.plot(list(sum_sq.keys()), list(sum_sq.values()),
        linestyle = '-', marker = 'H', color = 'g', 
         markersize = 10, markerfacecolor = 'red')
         
```
