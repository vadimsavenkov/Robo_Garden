## Combining and Merging

```python
d1 = {'key':['a','b','c','d','e'], 'A1':range(5), 'B1':range(5,10)}
d2 = {'key':['a','b','c'], 'A2':range(3), 'B2':range(3,6)}
```

```python
d1
```

```python
d2
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = 'inner', on = 'key') #merge using inner
```

```python
pd.merge(df1, df2, how = 'outer', on = 'key') #merge using outer
```

```python
pd.merge(df1, df2, how = 'left', on = 'key') #merge using left
```

```python
pd.merge(df1, df2, how = 'right', on = 'key') #merge using right
```

```python
left = pd.DataFrame({'key1': ['a', 'a', 'b', 'c'],
                     'key2': ['a', 'b', 'a', 'b'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['a', 'b', 'b', 'c'],
                      'key2': ['a', 'b', 'a', 'a'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how = 'inner', on = ['key1','key2'])
```

```python
pd.merge(left, right, how = 'outer', on = ['key1','key2'])
```

```python
pd.merge(left, right, how = 'left', on = ['key1','key2'])
```

```python
pd.merge(left, right, how = 'right', on = ['key1','key2'])
```
